## 03-1 불자료형과 if 조건문

### 불 자료형
- `==` : 같다
- `!=` : 다르다
- `<`  : 작다
- `>`  : 크다
- `<=` : 작거나 같다
- `>=` : 크거나 같다

````python
# 숫자 비교
print(10 == 100)
print(10 != 100)
print(10 < 100)
print(10 > 100)
print(10 <= 100)
print(10 >= 100)
````

- 한글은 사전 순서 (가나다순) 즉 사전 순으로 앞에 있는 것이 작은 값

````python
# 문자 비교
print("가방" == "가방")
print("가방" != "가방")
print("가방" < "하마")
print("가방" > "하마")
````

- not 연산자는 단항 연산자

- or 연산자는 양쪽 변의 값 중 하나만 참일 때만 True
#### 연산자
- 단항 연산자는 피연산자가 한 개; 예: 부호 연산자 -> 10을 음수로 만들기 위한 마이너스(-) 연산자
- 이항 연산자는 피연산자가 두 개 이상; 예: 계산식에 숫자 사이들 연산자

````python
print(not True)
print(not False)
````

````python
x = 10
under_20 = x < 20 # under_20 = (x < 20)
print(f"under_20 {under_20}")
print(f"not under_20 {not under_20}")
````

````python
print(True and True)
print(True and False)
print(False and True)
print(False and False) 
print(True or True)
print(True or False)
print(False or True)
print(False or False)
````

````python
 # 날짜/시간과 관련된 기능 모듈을 토해 가져옴
from datetime import datetime

# 현재 날짜/시간을 구합니다.
now = datetime.now()

# 출력합니다.
print(now.year, '년')
print(now.month, '월')
print(now.day, '일')
print(now.hour, '시')
print(now.minute, '분')
print(now.second, '초')

now1 = datetime.now()

print(f"{now1.year}년 {now1.month}월 {now1.day}일 {now1.hour}시 {now1.minute}분 {now1.second}초 ")
````

````python
# 계절 구별
from datetime import datetime

now = datetime.now()
if 3 <= now.month <= 5:
    print(f'이번 달은 {now.month}월로 봄입니다!')
elif 6 <= now.month <= 8:
    print(f'이번 달은 {now.month}월로 여름입니다!')
elif 9<= now.month <= 11:
    print(f'이번 달은 {now.month}월로 가을입니다!')
else:
    print(f'이번 달은 {now.month}월로 겨울입니다!')
````

- 조건문 작성시 이미 한번 비교한 것은 제외하고 작성

````python
# 변수를 선언합니다.
score = float(input("학점 입력> "))

# 조건문을 적용합니다.
if score == 4.5:
    print("신")
elif 4.2 <= score: # elif 4.2 <= score < 4.5 하지 말라고.. 혹은 4.2 <= score and score < 4.5 같은 거라던지
    print("교수님의 사랑")
elif 3.5 <= score:
    print("현 체제의 수호자")
elif 2.8 <= score:
    print("일반인")
elif 2.3 <= score:
    print("일탈을 꿈꾸는 소시민")
elif 1.75 <= score:
    print("오락문화의 선구자")
elif 1.0 <= score:
    print("불가촉천민")
elif 0.5 <= score:
    prin("t자벌레")
elif 0 < score:
    print("플랑크톤")
else:
    print("시대를 앞서가는 혁명의 씨앗")
````

### False 로 변환 되는 값

````python
print("if 조건문에 0 넣기")
if 0:
    print("0은 True")
else:
    print("0은 False")
    print()

print("if 조건문에 빈 문자열 넣기")
if "":
    print("빈문자열 True")
else:
    print("빈문자열 False")
````

## 04-1 리스트와 반복문
### 리스트
- 서로다른 자료형들 여러개를 하나의 변수에 저장

1) 대괄호 안에 음수를 넣어 뒤에서부터 요소를 선택할 수 있음
2) 리스트 접근 연산자를 다음과 같이 이중으로 사용 가능
3) 리스트 안에 리스트를 사용 -> 리스트 안에 데이터 프레임들 저장도 가능

````python
array = [273, 32, 103, "문자열", True, False]
print(array)
````

````python
for i in range(0, 6):
    print(f"{array[i]}")
````

````python
for i in range(-6, 0):
    print(f"{array[i]}")
````

### 리스트 연산하기: 연결(`+`), 반복(`*`), `len()`
- 단순 연산자는 변수 내용을 바꾸지 않는 비파괴적

````python
# 리스트를 선언합니다.
list_a = [1, 2, 3]
list_b = [4, 5, 6]

# 출력합니다.
print("# 리스트")
print("list_a =", list_a)
print("list_b =", list_b)
print()

 # 기본 연산자
print("# 리스트 기본 연산자")
print("list_a + list_b =", list_a + list_b)
print("list_a * 3 =", list_a * 3)
print()
# 함수
print("# 길이 구하기")
print("len(list_a) =", len(list_a))
print()

# 비파괴적인지 확인
print("# 비파괴적인지 확인")
print(list_a)
print(list_b)
````

### 리스트에 요소 추가하기: `append()`, `insert()`, `extend()`
- 변수가 변화됨

````python
# 리스트를 선언합니다.
list_a = [1, 2, 3]
print(f"list_a : {list_a}")
 # 리스트 뒤에 요소 추가하기
print("# 리스트 뒤에 요소 추가하기")
list_a.append(4)
list_a.append(5)
print(list_a)
print()

# 리스트 중간에 요소 추가하기
print("# 리스트 중간에 요소 추가하기")
list_a.insert(0, 10) # 전단계에서 추가
print(list_a)
list_a.insert(1, 11)
print(list_a)
print()

# extend() 함수
print("extend() 함수")
list_b = [4, 5, 6]
list_a.extend(list_b)
print(f"list_a : {list_a}")
````

### 리스트에 요소 제거하기

#### 인덱스로 제거하기: `del` 키워드, `pop()`

````python
list_a = [0, 1, 2, 3, 4, 5]
print("# 리스트의 요소 하나 제거하기,")
print()
# 제거 방법[1] - del 키워드
del list_a[1]
print("del list_a[l]:", list_a)
# 제거 방법[2] - pop()
list_a.pop(2)
print("pop(2):", list_a) #guide kor dg 데이터 분석 어떻게 시작하나요
````

### 리스트 정렬하기: `sort()`

### ⭐리스트 내부에 있는지 확인하기: `in`/`not in` 연산자
