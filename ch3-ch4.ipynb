## 03-1 불자료형과 if 조건문

### 불 자료형
- `==` : 같다
- `!=` : 다르다
- `<`  : 작다
- `>`  : 크다
- `<=` : 작거나 같다
- `>=` : 크거나 같다

````python
# 숫자 비교
print(10 == 100)
print(10 != 100)
print(10 < 100)
print(10 > 100)
print(10 <= 100)
print(10 >= 100)
````

- 한글은 사전 순서 (가나다순) 즉 사전 순으로 앞에 있는 것이 작은 값

````python
# 문자 비교
print("가방" == "가방")
print("가방" != "가방")
print("가방" < "하마")
print("가방" > "하마")
````

### 리스트 연산하기: 연결(`+`), 반복(`*`), `len()`
- 단순 연산자는 변수 내용을 바꾸지 않는 비파괴적

````python
# 리스트를 선언합니다.
list_a = [1, 2, 3]
list_b = [4, 5, 6]

# 출력합니다.
print("# 리스트")
print("list_a =", list_a)
print("list_b =", list_b)
print()

 # 기본 연산자
print("# 리스트 기본 연산자")
print("list_a + list_b =", list_a + list_b)
print("list_a * 3 =", list_a * 3)
print()
# 함수
print("# 길이 구하기")
print("len(list_a) =", len(list_a))
print()

# 비파괴적인지 확인
print("# 비파괴적인지 확인")
print(list_a)
print(list_b)
````

### 리스트에 요소 추가하기: `append()`, `insert()`, `extend()`
- 변수가 변화됨

````python
# 리스트를 선언합니다.
list_a = [1, 2, 3]
print(f"list_a : {list_a}")
 # 리스트 뒤에 요소 추가하기
print("# 리스트 뒤에 요소 추가하기")
list_a.append(4)
list_a.append(5)
print(list_a)
print()

# 리스트 중간에 요소 추가하기
print("# 리스트 중간에 요소 추가하기")
list_a.insert(0, 10) # 전단계에서 추가
print(list_a)
list_a.insert(1, 11)
print(list_a)
print()

# extend() 함수
print("extend() 함수")
list_b = [4, 5, 6]
list_a.extend(list_b)
print(f"list_a : {list_a}")
````

## 이터레이터
-  이터러블(iterable): ‘반복할 수 있는 것’
- 내부에 있는 요소들을 차례차례 꺼낼 수 있는 객체
    -  리스트, 딕셔너리, 문자열, 튜플
- `next()` 함수를 적용해 하나하나 꺼낼 수 있는 요소를 이터레이터 iterater라고 함

````python
numbers = [1, 2, 3, 4, 5, 6]
r_num = reversed(numbers)

print("reversed_numbers :", r_num)
print(next(r_num))
print(next(r_num))
print(next(r_num))
print(next(r_num))
print(next(r_num))
print(r_num)
````

- `reversed()` 함수의 리턴값이 바로 ‘reverseiterator’로 ‘이터레이터’임
- 이와 같은 이터레이터는 반복문의 매개변수로 전달할 수 있으며, 현재 코드처럼 `next()` 함수로 
내부의 요소를 하나하나 꺼낼 수 있음

- 이터레이터함수들이 리스트를 바로 리턴해 주지 않고 이터레이터를 리턴하는 이유는 메모리의 효율성을 위함
- 1만 개의 요소가 들어 있는 리스트를 복제한 뒤 뒤집어서 리턴하는 것보다 기존에 있던 리스트를 활용해서 작업하는 것이 훨씬 효율적이라고 판단하기 때문
